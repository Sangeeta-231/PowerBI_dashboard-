```python
numerical_columns = dataset.select_dtypes(include=['float64', 'int64']).columns

# Detecting and handling outliers using the IQR method
outlier_info = {}
cleaned_data = dataset.copy()  # Create a copy of the dataset to handle outliers

for col in numerical_columns:
    Q1 = dataset[col].quantile(0.25)  # First quartile
    Q3 = dataset[col].quantile(0.75)  # Third quartile
    IQR = Q3 - Q1  # Interquartile range
    
    # Calculate bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = dataset[(dataset[col] < lower_bound) | (dataset[col] > upper_bound)]
    outlier_info[col] = len(outliers)
    
    # Handle outliers by capping them to bounds
    cleaned_data[col] = dataset[col].clip(lower=lower_bound, upper=upper_bound)

# Display outlier summary and first few rows of cleaned data
print("Outliers detected per column:")
for col, count in outlier_info.items():
    print(f"{col}: {count} outliers")

print("\nCleaned dataset (first 5 rows):")
print(cleaned_data.head())
```
